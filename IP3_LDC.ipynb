# Rename the columns

```python
df.rename({'Date/Time': 'date_time', 'Temp (C)': 'temp_c', 'Dew Point Temp (C)': 'dew_point', 'Rel Hum (%)': 'relative_humidity',
           'Wind Spd (km/h)': 'wind_speed', 'Visibility (km)': 'visibility', 'Stn Press (kPa)': 'pressure', 'Weather': 'weather' 
          }, axis=1, inplace=True)
df
```

# Add a column that contains the temperature in Farenheit

```python
df = df.assign(temp_f = df.temp_c*(9/5)+32)
df
```

# Which month of the year has temp_f more than 90, humidity less than 50 , dew point more than 17, and clear weather? 

```python
df.loc[(df.temp_f > 90) & (df.relative_humidity < 50) & (df.dew_point > 17) & (df.weather == 'Clear')].head(1)
```

# Extract this data and save it to a new data frame

```python
df1 = df.loc[(df.temp_f > 90) & (df.relative_humidity < 50) & (df.dew_point > 17) & (df.weather == 'Clear')].head(1)
df1.to_csv("month", index = False)
```

# Extract the month information from the date_time column and add a month column at the data frame.

```python
df['month'] = df['date_time'].dt.month
df.head() # Here we can see a month column has been added
```
